```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")  # download the model and return as object ready for use
```

```python
word_vectors=model

# Let us look how the vector embedding of a word looks like
print(word_vectors['computer'])  
```

```python
print(word_vectors.most_similar(positive=['boy', 'girl'], negative=['men'], topn=10))
```

```python
print(word_vectors.most_similar(positive=['boyfriend','girlfriend','father'],negative=['father'],topn=50))
```

```python
print(word_vectors.similarity('boy','penis'))
print(word_vectors.similarity('boy','pen'))
print(word_vectors.similarity('boyfriend','girlfriend'))
```
